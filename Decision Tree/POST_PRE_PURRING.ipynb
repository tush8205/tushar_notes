Decision Tree Classifier Implementation with Post Purring

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset.DESCR
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
#Independent amd depeendt feeature 
X=df.iloc[:,:-1]
y=dataset.target
```

```python
X,y
```

```python
#Train test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
##Post Prunning 
treeClassifier=DecisionTreeClassifier()
```

```python
treeClassifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

```python
##Post Prunning with max_depth=2
treeClassifier=DecisionTreeClassifier(max_depth=2)##creteria='entropy'
treeClassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

```python
##Accuary score, Classifiction Report-- find out 
```

```python
y_pred=treeClassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
score=accuracy_score(y_pred,y_test)
score
```

```python
Report=classification_report(y_pred,y_test)
print(Report)
```

Decision Tree Classifier Implementation with Pre Purring and hyperparameter tunning for huge data 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
'criterion':['gini','entropy', 'log_loss'],
'splitter':['best','random'],
'max_depth':[1,2,3,4,5], 
'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeClassifier=DecisionTreeClassifier()
clf=GridSearchCV(treeClassifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred=clf.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
classReport=classification_report(y_pred,y_test)
print(classReport)
```
