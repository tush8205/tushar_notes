LINER INTERPPLOTION

```python
import numpy as np
```

```python
x= np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
x_new=np.linspace(1,5,10) 
y_interp= np.interp(x_new,x,y)
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

Cubic interpolation woth scipy

```python

import numpy as np
x= np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
pip install scipy
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind='cubic')
```

```python
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
print(y_interp)
```

```python

import matplotlib.pyplot as plt
plt.scatter(x_new,y_interp)
```

```python
#Polynomial Function
import matplotlib.pyplot as plt
```

```python

import numpy as np
x= np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])

```

```python
p = np.polyfit(x,y,4)
```

```python
x_new= np.linspace(1,5,10)
y_interp=np.polyval(p,x_new)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

```python
plt.scatter(x_new,y_interp)
plt.xlabel("x Axis")
plt.ylabel('Y axis ')
plt.title('Cubic ')
```
