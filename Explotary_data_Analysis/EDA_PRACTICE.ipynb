```python
df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Day']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Day']=df['Day'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.info()
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_Min']=df['Dep_Min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_Min']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Min']=df['Arrival_Min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Duration'].isnull().sum()
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_hour']
```

```python
df['Duration_Min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df['Duration_Min']
```

```python
df['Duration_Min']=df['Duration_Min'].replace(np.NaN,0)
df['Duration_Min']
```

```python
df['Duration_Min'].isnull().sum()
```

```python
df['Duration_hour']=df['Duration_hour'].astype(int)
```

```python
df['Duration_hour'].str.isnumeric().sum()
```

```python
df[~df['Duration_hour'].str.isnumeric()]
```

```python
df['Duration']=df['Duration'].replace(['5m'],'5')
df['Duration_hour']=df['Duration_hour'].replace(['5m'],'5')
```

```python
df['Duration_hour']=df['Duration_hour'].astype(int)
df['Duration']=df['Duration'].astype(int)
```

```python
df['Duration_Min']=df['Duration_Min'].astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```
