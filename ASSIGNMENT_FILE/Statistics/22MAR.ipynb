```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['location.name'].unique()
```

```python
df['location.name'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df['location.name'],data=df,palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['location.latitude'],kde=True,palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(df['location.name'].value_counts(),labels=df['location.name'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['location.depth'],y=df['location.distance'],hue=df['location.name'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df['location.longitude'],y=df['location.latitude'],palette='hls')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df['location.longitude'],y=df['location.latitude'],hue=df['impact.gap'])
plt.show()
```
