Q2.: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

ANS:A Scatter plot is a type of plot or mathermatical diagram using Cartensian coordinates to display values for typically two variable for a set of data.The data are displayed as a collection of points, each having  the value of one  variable  detemining the position on the horizontal axis and the value  of the   other variable determining the position  on the vartical axis.


```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.scatter(x,y)
```

Q3.Why is the subplot() function used? Draw four line plots using the subplot() function.
ANS:Subplot() which acts as a utility wrapper and helps in creating common layouts of subplots , including the enclosing figure object, in a single call.
as simple , provides a way to plot multiple plots on a single figure.

```python
#1.
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y,color="yellow",marker="*")
plt.subplot() 
```

Q4.What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
ANS:A Barplot is one of the most common types of graphic.It shoes the relationship between a numeric and caregoric variable.
Bar charts should be used when you are showing segements  of informantion.Vertical bar charts are useful to compare different categorical or discrete variables , such  as age group,classes, schools.


```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar Plot")
plt.bar(company,profit)
```

Q5.What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
ANS:The method to summarize a set of data is measured using an inteval scale is called Box Plot.These are maximum used for data analysis.They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
plt.show()
```
