Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
plot1=sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri,hue="event")
plot1
```

```python
plot2=sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri,hue="region",style='event')
plot2
```

Q4.: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 



```python
import seaborn as sns
diamond=sns.load_dataset("diamonds")
diamond
```

```python
diamond=sns.displot(diamond,x='price',hue='cut')
diamond
```

Q5.Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset. 

```python
import seaborn as sns
iris=sns.load_dataset('iris')
iris
```

Q6.Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set()
flight=sns.load_dataset('flights')
flight=flight.pivot("month","year","passengers")
ax=sns.heatmap(flight)
plt.show()
```
