Q1. What is Flask Framework? What are the advantages of Flask Framework?
ANS:Flask ia a micro web framework written in python. It is classified as a microframwork because it does not require particular tools or libraries. It has no database abstration layer, from validation, or any other componets where pre-existing third party libraries provide common function.
Advantages:
1.It is a lightweight framwork that offers hassle free development.
2.Provide flexibility to the developer to experiment with their modules or architecture.
3.It is suitable for small project.
4.Easily scalable for the application
5.Support for secure cookies
6.Provide strong WSGI support.

Q5.What function is used in Flask for URL Building?Write a Python code to demonstrate the working of the
url_for() function.
ANS:The url_for() function is used to build a URL to the specific function dynamically. The first argument is the name of the specified function, and then we can pass any number of keyword argument corresponding to the variable part of the URL.

from flask import Flask,url_for
app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the Homepage'

@app.route('/user/<username>')
def show_user_profile(username):
    return f'Hello,{username}'

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f'This is post number {post_id}'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('show_user_profile',username='tushar'))
    print(url_for('show_post',post_id=5))
